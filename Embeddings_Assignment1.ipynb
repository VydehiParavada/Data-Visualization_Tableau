```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large", api_key=openai_api_key)

embed1 = embeddings.embed_query("Hello How Are You")
embed2 = embeddings.embed_query("How is the weather?")
embed3 = embeddings.embed_query("The class is interesting")

print(f"Printout for the embedding model:{embed1}")
print(f"Printout for the embedding model:{embed2}")
print(f"Printout for the embedding model:{embed3}")
```
